```python
import opensmile
import pandas as pd

smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.GeMAPSv01b,
    feature_level=opensmile.FeatureLevel.Functionals,
)
df = smile.process_file('~/Downloads/audio_file.wav')
```

```python
#measuring F0 relative to 27.5 Hz, it allows for a standardized measure of the speaker's voice pitch that can be used across different recordings and individuals.
avg_f0_semitone_score = df['F0semitoneFrom27.5Hz_sma3nz_amean'] #F0semitoneFrom27.5Hz_sma3nz_amean is a feature that measures the average pitch frequency of a person's voice. In the context of Alzheimer's disease, changes in speech patterns and characteristics are often observed, including changes in pitch, volume, and intonation. Research suggests that individuals with Alzheimer's disease may have a lower average pitch frequency than those without the disease, which could be indicative of vocal atrophy and muscle weakness.
f0_meanRisingSlope = df['F0semitoneFrom27.5Hz_sma3nz_meanRisingSlope'] #The rising slope measures how fast the F0 contour increases over time, and the mean rising slope measures the average of all rising slopes in the signal. This feature can provide information about the pitch variability and pitch instability in speech, which can be affected by neurological disorders such as Alzheimer's disease.
f0_stddevRisingSlope = df['F0semitoneFrom27.5Hz_sma3nz_stddevNorm'] #The standard deviation of F0 normalized by the mean, as captured by F0semitoneFrom27.5Hz_sma3nz_stddevNorm, is a commonly used measure of F0 variability, or jitter, and has been found to be significantly increased in individuals with Alzheimer's disease compared to healthy controls
```

```python
import os
import glob
import opensmile
import pandas as pd

# Set up OpenSmile
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.GeMAPSv01b,
    feature_level=opensmile.FeatureLevel.Functionals,
)

# Set up an empty DataFrame to store the features
all_features_df = pd.DataFrame()

# Loop through all .wav files in the downloads folder
for audio_file_path in glob.glob(os.path.expanduser('~/Downloads/*.wav')):
    # Extract features for the audio file using OpenSmile
    features = smile.process_file(audio_file_path)
    # Add the features to the DataFrame
    all_features_df = all_features_df.append(features, ignore_index=False)

# View the DataFrame
all_features_df = all_features_df.transpose()
all_features_df

```
